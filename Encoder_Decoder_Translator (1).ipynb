```python
voc1size=len(tokenizer1.word_index)
voc2size=len(tokenizer2.word_index)
```

```python
from tensorflow import keras
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
```

```python
from tensorflow.keras.models import Model
```

```python
encoder_inputs = Input(shape=(len(sequences1[0]),), name="encoder_inputs")
```

```python
encoder_embedding = Embedding(
    input_dim=voc1size,
    output_dim=256,
    mask_zero=True
)(encoder_inputs)
```

```python
# Layer 1
x, state_h1, state_c1 = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(encoder_embedding)

# Layer 2
x, state_h2, state_c2 = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(x)

# Layer 3
x, state_h3, state_c3 = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(x)

# Layer 4
encoder_outputs, state_h4, state_c4 = LSTM(
    200,
    return_sequences=False,
    return_state=True
)(x)
```

```python
decoder_inputs = Input(shape=(None,), name="decoder_inputs")
```

```python
decoder_embedding = Embedding(
    input_dim=voc2size,
    output_dim=256,
    mask_zero=True
)(decoder_inputs)
```

```python
# Layer 1
x, _, _ = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(
    decoder_embedding,
    initial_state=[state_h1, state_c1]
)

# Layer 2
x, _, _ = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(
    x,
    initial_state=[state_h2, state_c2]
)

# Layer 3
x, _, _ = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(
    x,
    initial_state=[state_h3, state_c3]
)

# Layer 4
decoder_outputs, _, _ = LSTM(
    200,
    return_sequences=True,
    return_state=True
)(
    x,
    initial_state=[state_h4, state_c4]
)
```

```python
decoder_outputs = Dense(
    voc2size,
    activation="softmax"
)(decoder_outputs)
```

```python
model = Model(
    [encoder_inputs, decoder_inputs],
    decoder_outputs
)
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```
